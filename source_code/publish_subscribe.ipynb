# 2. 发布者-订阅者模式

Publisher为诗歌发布中心，Reader为订阅相应诗人或相应的关键词或诗人+关键词的诗歌的人。可实现订阅，退订。当Reader订阅后，Publisher会及时通知Reader其订阅的内容。


**Publisher（发布者）**

    - 保存订阅的读者
    - 记录所有诗人的诗歌的总数
    - 发布订阅者订阅的诗歌
 
   
**Reader（订阅者）**

    - 订阅/取消订阅 发布者（诗人、关键词、诗人+关键词）
    - 保存发布者发来的诗歌
    - 保存所有消息

```python
# Base发布者 & 发布者
class BasePublisher(object):
    def __init__(self,name,authors_list, word_list,syn):
        self.authors = authors_list
        self.readers = {}
        self.name=name
        self.totalNumber={}
        self.word_list = word_list
        self.syn = syn
        for i in self.authors:
            self.totalNumber[i] = 0
            self.readers[i]=[]
        self.poems = []
        
    def subscribeReader(self, author, reader, sentence):
        if author not in self.readers:
            self.readers[author] = []
        self.readers[author].append(reader)
        sentences = sentence.split(' ')
        for word in sentences:
            if word not in self.readers:
                self.readers[word] = []
            self.readers[word].append(reader)
        self.notifyReader(author,reader,sentence)
        return self

    def unsubscribeReader(self, author, reader,sentence):
        self.readers[author].remove(reader)
        sentences = sentence.split(' ')
        for word in sentences:
            self.readers[word].remove(reader)
        return self

    def notifyReader(self,author,reader,sentence):
        if (len(sentence)> 0 and len(author)>0):
            tmp = author + " "+sentence
            sentences = sentence.split(' ')

            intersection = set(self.word_list[sentences[0]])
            cnt = 0
            for word in sentences:
                intersection = intersection & set(self.word_list[word])
            inter = list(intersection)
            for i in self.poems:
                if author == i.name and i.id in inter:
                    reader.receivePoem(self,i,tmp)
#             print("im back1")
            tmp_inter = set(self.word_list[sentences[0]])      
            for word in sentences:
#                 print("hello you are in1")
                union = set()
                if word in self.syn:
                    sy_word = self.syn[word]
                    union = set(self.word_list[word])
#                     print("inin1")
                    for s in sy_word:
#                         print("checking1")
                        union = union | set(self.word_list[s])        
                    if(cnt==0):
                        tmp_inter = union
                    else:
                        tmp_inter = union & tmp_inter
                else:
                    tmp_inter = tmp_inter & set(self.word_list[word])
                cnt+=1    
            tmp_inter = tmp_inter-intersection
            inter = list(tmp_inter)
            for i in self.poems:
                if author == i.name and i.id in inter:
                    reader.receivePoem(self,i,tmp)
        elif (len(sentence)>0 and len(author) == 0):
            tmp = sentence
            sentences = sentence.split(' ')
            intersection = set(self.word_list[sentences[0]])
            cnt = 0
            for word in sentences:
                intersection = intersection & set(self.word_list[word])
            inter = list(intersection)
            for i in self.poems:
                if i.id in inter:
                    reader.receivePoem(self,i,tmp)
            tmp_inter = set(self.word_list[sentences[0]])      
            for word in sentences:
                if word in self.syn:
                    sy_word = self.syn[word]
                    union = set(self.word_list[word])
                    for s in sy_word:
                        union = union | set(self.word_list[s])
                    if(cnt==0):
                        tmp_inter = union
                    else:
                        tmp_inter = union & tmp_inter
                else:
                    tmp_inter = tmp_inter & set(self.word_list[word])
                cnt+=1    
            tmp_inter = tmp_inter-intersection
            inter = list(tmp_inter)
            for i in self.poems:
                if i.id in inter:
                    reader.receivePoem(self,i,tmp)
        elif (len(sentence)==0 and len(author) > 0):
            for i in self.poems:
                if author == i.name:
                     reader.receivePoem(self,i,author)
#         news = '{}的诗{}'.format(author,poem)
#         self.poems.append(news)
#             for reader in self.readers[author]:
#                 reader.receivePoem(self, poem, author)
    
    def __str__(self):
        readers = f'Subscribed Reader: {set([reader.name for author in self.authors for reader in self.readers[author]])}\n'
        poems = f'发出的诗歌:{[(key,value) for key,value in self.totalNumber.items()]}\n'
        return readers + poems

    def publish_poem(self,new_df):
    # print()
        for index, row in new_df.iterrows():
            author = row['Author']
            title = row['Title']
            content = row['Content']
            idd = row['Poem_id']
            if(author in self.authors):
                self.totalNumber[author] += 1
        #     print(author,title,content)
            tmp1 = poem(title,author,idd, content)
            self.poems.append(tmp1)
#         global all_poem
#         all_poem = self.poems.copy()


class Publisher(BasePublisher):
    def __init__(self,name,authors_list, word_list,syn):
        BasePublisher.__init__(self,name,authors_list, word_list,syn)

#         #notify part havent settle
#     def notifyReader(self):
# #         BasePublisher.notifyReader()
#         pass
            
    def publish_poem(self, new_df):
        poem = BasePublisher.publish_poem(self,new_df)
        
```
